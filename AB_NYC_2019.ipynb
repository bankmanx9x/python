```python
import pandas as pd
import numpy as np

link = '/content/AB_NYC_2019.csv'
df = pd.read_csv(link)
df.head()
```

```python
df = df[['neighbourhood_group', 'room_type', 'price']]
df.head()
```

```python
d = df.groupby(['neighbourhood_group', 'room_type'])['price'].describe()[['mean', 'count', '25%', '50%', '75%']]
d.to_csv('sample.csv')
```

```python
gb = pd.read_csv('/content/airbnb_gb_nb_type.csv')

gb['iqr'] = gb['75%'] - gb['25%']
gb.head(10)
```

```python
gb['lower_range'] = gb['25%'] - 1.5*gb['iqr']
gb['upper_range'] = gb['75%'] + 1.5*gb['iqr']
gb[['neighbourhood_group', 'room_type', 'lower_range', '25%', '50%', '75%','upper_range']]
```
